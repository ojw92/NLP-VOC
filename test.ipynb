### Test on Today's Dataset

```python
from bertfunc import data_concat, pred_day

# Make predictions on all models for today's data

berttype = {'s22'   : 'BertS22',
            's23'   : 'BertS23',
            'fold'  : 'BertFold',
            'flip'  : 'BertFlip',
            #'watch' : 'BertWatch',
            'tab'   : 'BertTab'  }
voc_today = pd.read_csv('20230914_test 1.csv', index_col=0, delimiter=",").drop_duplicates()
pd.set_option('display.max_rows', None,
              'display.max_columns', None,
              'display.precision', 3,)

for modeltype in list(berttype.keys()):
  model_2 = transformers.BertForSequenceClassification.from_pretrained(berttype[modeltype])
  model_2.to('cuda')

  voc_today_model = data_concat(voc_today[voc_today['Model']== modeltype ].loc[: , 'Title':'Class'])

  print(f'{modeltype} VOC')
  print(pred_day(voc_today_model, model_2), '\n')
```
